```python
tab8tsv = pd.read_csv('table8.tsv', delimiter='\t')
tab8tsv
```

```python
acc = {}
for r in tab8tsv.to_dict(orient="records"):
    acc[('mBERTlarge', r['Language'])] = {
        'Zero-shot Performance': f"{r['Zero-Shot Performance (Mean)']:#.4} $\pm$ {r['Zero-Shot Performance (Stdev)']:#.4}",
        'Translate-Test': f"{r['Translate-Test (Mean)']:#.4} $\pm$ {r['Translate-Test (Stdev)']:#.4}",
        'Cross-Lingual Transfer Gap': f"{r['Cross-Lingual Gap']:#.4}",
        'Concept Shift Gap': f"{r['Concept Shift Gap']:#.4}",
    }
    
df = pd.DataFrame.from_dict(acc, orient='index')
# df.index.set_names('Model', level=0, inplace=True)
# df.index.set_names('Language', level=1, inplace=True)
dfs = df.style
def index_formatter(s):
    replacer = {
        'en_dev': 'en$_{dev}$',
        'mBERTlarge': 'mBERT$_{large}$'
    }
    if s in replacer:
        return replacer[s]
    return s
dfs.format_index(index_formatter)
dfs.format_index(lambda s: f"\\textbf{{{s}}}", axis=1)
```
